```python
Figures, Axes = plt.subplots(figsize=(140,60))

c=np.random.randint(1,10,size=len(data))
s=np.random.randint(10,220,size=len(data))

Axes.scatter(data['Count'],data['Temperature'],s=s,c=c)
```

```python
Labels={"Spring", "Summer", "Autumn", "Winter"}
Counts=Spring['Count'].sum(), Summer['Count'].sum(), Autumn['Count'].sum(), Winter['Count'].sum()
```

```python
Figures_2, Axes_2= plt.subplots(figsize=(14,8))

Axes_2.pie(Counts,labels=Labels,autopct='%1.1f%%',explode=[0.1,0,0.1,0])
Axes_2.legend()
plt.show()
```
